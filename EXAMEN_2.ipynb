Crea un grafico donde se pueda ver la comparacion de ingresos de adultos por intervalo de  años: 2015, 2016,2017

```python
df.info()
```

```python
df.sort_values(by="adults",inplace=True)
```

```python
df.head()
```

```python
df_pl=df.groupby(["adults"])[["arrival_date_year", "arrival_date_day_of_month", "adr", "reservation_status_date"]].max().reset_index()
df_pl
```

```python
plt.plot(df_pl['adults'],df_pl['arrival_date_day_of_month'], 'rP:', label='Adultos vs Fecha de llegada')
plt.annotate('Este valor indica la falta de pasajeros', xy=(1,30), xytext=(5,28), arrowprops=dict(facecolor='black', shrink=0.01))
plt.xlabel("Adultos")
plt.ylabel("Fecha de llegada")
plt.legend()
plt.show()
```

En este grafico se muestra la evulucion de llegada de pasajeros desde el año 2015 hsta el periodo 2017.
